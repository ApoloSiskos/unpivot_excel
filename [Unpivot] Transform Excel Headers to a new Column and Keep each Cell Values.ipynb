```python
import pandas as pd
import numpy as np
from datetime import datetime
from dateutil.relativedelta import relativedelta

def get_csv():
    #Read csv file
    df = pd.read_excel('12711-0002.xls')
    
    #Fill in the empty cells with the value of the above cell
    df = df.fillna(method='ffill')
    
    #Convert column headers into index
    df = (df.set_index(['Country', 'Gender', 'Arr-Dep'])
             .stack()
             .reset_index(name='Value')
             .rename(columns={'level_3':'Year'}))    
    return df

get_csv()
```
