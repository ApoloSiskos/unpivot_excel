```python
import pandas as pd
import numpy as np
from datetime import datetime
from dateutil.relativedelta import relativedelta

def get_csv():
    #Read csv file
    df = pd.read_excel('12711-0002.xls')
    
    #Version 1 
    df1 = (df.ffill()
             .melt(id_vars=['Country','Gender','Arr-Dep'],var_name="Year",value_name='Value')
    )    
    
    '''
    #Full Version with sorting
    df2 = (df.ffill()
         .melt(id_vars=['Country','Gender','Arr-Dep'],var_name="Year", value_name='Value')
         .sort_values(['Country', 'Gender','Arr-Dep'])
         .reset_index(drop=True))
    '''
    
    # Create a Pandas Excel writer using XlsxWriter as the engine.
    writer = pd.ExcelWriter('new.xlsx', engine='xlsxwriter')

    # Convert the dataframe to an XlsxWriter Excel object.
    df1.to_excel(writer, sheet_name='Sheet1',header=True, index=False)

    # Close the Pandas Excel writer and output the Excel file.
    writer.save()
    
    return df1

get_csv()
```
